```python
# Handling missing values

median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
print(f"Missing age values filled with median value: {median_age}")
```

Question 1

```python
# 1
df['fare_per_year'] = df['Fare'] / df['Age']

# 2
high_fare_age = df[df['fare_per_year'] > 5]

# 3
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# 4
result_q1 = high_fare_age_srt[['Name', 'fare_per_year']]

# 5
print("Top fare per year:")
print(result_q1.head())
```

3.3 Exercise

Question 1

```python
# 1
total_fare = df['Fare'].sum()

# 2
fare_c1 = df[df['Pclass'] == 1]['Fare'].sum()

# 3
fare_c2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_c3 = df[df['Pclass'] == 3]['Fare'].sum()

# 4
fare_totals = [fare_c1, fare_c2, fare_c3]

# 5
fare_proportions = [fare / total_fare for fare in fare_totals]

classes = ['First class', 'Second class', 'Third class']
for cls, prop in zip(classes, fare_proportions):
    print(f"{cls} fare percentage: {prop:.2%}")
```

Question 2

```python
# 1
def get_age_group(age):
    if age < 18:
        return 'child'
    elif 18 <= age <= 64:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(get_age_group)

# 2
total_passengers = len(df)

# 3
group_counts = df['age_group'].value_counts()

# 4
passengers_proportions = (group_counts / total_passengers)

# 5
proportion_percentage = passengers_proportions * 100
print("Percentages of age groups from total passengers:")
print(proportion_percentage)
```
