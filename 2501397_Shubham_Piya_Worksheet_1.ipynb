Probem - 4: Matrix Operations:

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

```python
# 1.1

A_inv = np.linalg.inv(A)
result = A @ A_inv
result = np.round(result).astype(int)

identity = np.eye(2, dtype = int)

print("A * A^-1 =\n", result, "\nI =\n", identity)
if (np.array_equal(result, identity)):
  print("A * A^-1 = I")
```

```python
# 1.2

AB = A @ B
BA = B @ A

print("AB =\n", AB, "\nBA =\n", BA)

print("A * B = B * A:", np.array_equal(AB, BA))
```

```python
# 1.3

LHS = (A @ B).T
RHS = B.T @ A.T

print("LHS =\n", LHS, "\nRHS =\n", RHS)
if (np.array_equal(LHS, RHS)):
  print("(A.B)T (LHS) = (B)T.(A)T (RHS)")
```

```python
# 2

"""
Equations:
2x - 3y + z = -1
 x - y + 2z = -3
3x + y -  z =  9

"""

A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3,  1, -1]], dtype=float)

B = np.array([-1, -3, 9], dtype=float)

# AX = B
# X = A^-1 . B
# Finding X using np.linalg.inv()

A_inverse = np.linalg.inv(A)
X = A_inverse @ B
print("Solving using inverse matrix method:\n", X)
```

```python
# 3

# Finding X using np.linalg.solve()
X_solve = np.linalg.solve(A, B)
print("Solving using np.linalg.sovle():\n", X_solve)
```

```python
import time

# python lists
list1 = list(range(1000000))
list2 = list(range(1000000))

# numpy arrays
array1 = np.arange(1000000)
array2 = np.arange(1000000)
```

```python
# 1

# python lists

start = time.time()
list_add = [list1[i] + list2[i] for i in range(1000000)]
end = time.time()
print("Time for elementwise addition of python lists:", end - start, "s")

# numpy arrays

start = time.time()
array_add = array1 + array2
end = time.time()

print("Time for elementwise addition of numpy arrays:", end - start, "s")
```

```python
# 2

# python lists

start = time.time()
list_multiply = [list1[i] * list2[i] for i in range(1000000)]
end = time.time()
print("Time for elementwise multiplication of python lists:", end - start, "s")

# numpy arrays

start = time.time()
array_multiply = array1 * array2
end = time.time()

print("Time for elementwise multiplication of numpy arrays:", end - start, "s")
```

```python
# 3

# python lists

start = time.time()
list_dot = sum(list1[i] * list2[i] for i in range(1000000))
end = time.time()
print("Time for dot product of python lists:", end - start, "s")

# numpy arrays

start = time.time()
array_dot = np.dot(array1, array2)
end = time.time()

print("Time for dot product of numpy arrays:", end - start, "s")
```

```python
# 4

# python matrices
python1 = [[1 for _ in range(1000)] for _ in range(1000)]
python2 = [[1 for _ in range(1000)] for _ in range(1000)]

start = time.time()
python3 = [[sum(python1[i][k] * python2[k][j] for k in range(1000))
           for j in range(1000)] for i in range(1000)]
end = time.time()

print("Time for python matrix multiplication:", end - start, "s")

# numpy matrices
np1 = np.ones((1000, 1000))
np2 = np.ones((1000, 1000))

start = time.time()
np3 = np1 @ np2
end = time.time()

print("Time for numpy matrix multiplication:", end - start, "s")
```
